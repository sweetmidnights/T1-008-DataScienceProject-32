```r
library(tidyverse)
library(dplyr)
library(readr)
library(tidymodels)
library(repr)
library(rvest)
library(stringr)
library(RColorBrewer)
library(janitor)
```

```r
url <- "https://raw.githubusercontent.com/sweetmidnights/T1-008-DataScienceProject-32/main/crop_production.csv"
dataset <- read_csv(url)

dataset
```

```r
dataset_split <- initial_split(dataset, prop = 0.75) 
dataset_train <- training(dataset_split)
dataset_test <- testing(dataset_split)
dataset_train
dataset_test
```

```r
filter_dataset_train <- filter(dataset_train, SUBJECT == "RICE" & Value > 10000)
```

```r
plot_student_performance <- ggplot(filter_dataset_train , aes(x = TIME, y = Value, color = SUBJECT)) +
                   geom_point() +
                   xlab("Time (years)") +
                   ylab("Crop's Value") + 
                   labs(color = 'Crop Name') +
                   theme(text = element_text(size = 12))+
                   ggtitle("Crop Proportion over time") #title for the plot

plot_student_performance
```
